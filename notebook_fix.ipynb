```python
from datetime import date
def usiaPegawai(x):
    today=date.today()
    return today.year - x.year - ((today.month, today.day) < (x.month, x.day))
```

```python
def getNilaiKinerja(x,value):
    temp= x/value*100
    if temp!=100:
        return False
    else:
        return True
```

```python
target_kinerja={'1.0':7200,'2.0':6800,'3.0':7200,'4.0':6800,'5.0':5400,'6.0':7200}
temp=0
for key,value in target_kinerja.items():
    temp+=value
pegawai['Kinerja On Target']=pegawai['poin aktivitas diakui'].apply(lambda x:getNilaiKinerja(x,temp))
pegawai
```

```python
def createGraph(x,indikator):
    names = list(x.keys())
    values = list(x.values())
    temp1=len(values[0])
    temp2=len(values[1])
    plt.pie([temp1,temp2],labels=names,startangle=90, shadow=True,explode=(0.1, 0.1), autopct='%1.2f%%')
    plt.title('Diagram Disiplin Berdasarkan Indikator {}'.format(indikator))
    plt.axis('equal')
    plt.show()
```

Menimbang faktor pendidikan masing-masing pegawai

```python
pendidikan_tl1_nonTarget=kinerja_tl1_nonTarget['tingkat pendidikan'].unique()
kamus_pendidikan={}
for item in pendidikan_tl1_nonTarget:
    kamus_pendidikan[item]=0
kamus_pendidikan
```

```python
def getJumlahPendidikan(dataframe,kamus):
    for index,row in dataframe.iterrows():
        for key,value in kamus.items():
            if key==row['tingkat pendidikan']:
                kamus[key]=value+1
    return kamus
```

```python
def createGraph(x,indikator):
    names = list(x.keys())
    values = list(x.values())
    plt.pie(values,labels=names,startangle=90, shadow=True, autopct='%1.2f%%')
    plt.title('Diagram Pendidikan pegawai Berdasarkan Indikator {}'.format(indikator))
    plt.axis('equal')
    plt.show()
```

```python
createGraph(getJumlahPendidikan(kinerja_tl1_nonTarget,kamus_pendidikan),'Tidak disiplin presensi TL1 dan Non Target Kinerja')
```

```python
getJumlahPendidikan(kinerja_tl1_nonTarget,kamus_pendidikan)
```
