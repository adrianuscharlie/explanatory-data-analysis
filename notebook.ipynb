Kami ingin mengangkat permasalahan seberapa banyak Pegawai ASN di Kota Yogyakarta yang mendapat pengurangan TPP ASN, berapa persen jumlah pengurangan berdasarkan aspek presensi pegawai dengan indikator presensi tepat waktu dan pulang sebelum waktu kantor dan unit kerja mana yang mendapatkan rerata pengurangan  TPP ASN paling besar sesuai dengan aspek sebelumnya.

Import semua libraries yang dibutuhkan.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Import semua data excel menjadi dataframe.

```python
pegawai=pd.read_excel('Data Pegawai.xlsx')
kinerja=pd.read_excel('E-Kinerja.xlsx')
office=pd.read_excel('E-Office.xlsx')
presensi=pd.read_excel('Presensi.xlsx')
```

Mencari informasi terhadap semua atribut/value yang ada di data pegawai.

Karena terlalu banyak informasi yang kurang diperlukan dalam melakukan analisis, maka melakukan drop beberapa kolom data.

```python
delete_column=['pegawai_propinsi_nama','pegawai_kecamatan_nama','pegawai_kabupaten_nama','pegawai_kelurahan_nama','nama pendidikan','pangkat','CPNS TMT']
for column in delete_column:
    del pegawai[column]
pegawai
```

```python
pegawai['nama unit kerja'].unique()
```

```python
from datetime import date
def usiaPegawai(x):
    today=date.today()
    return today.year - x.year - ((today.month, today.day) < (x.month, x.day))
```

```python
pegawai['usia']=pegawai['pegawai_tgl_lahir'].apply(usiaPegawai)
pegawai
```

Menampung ID Pegawai yang akan digunakan untuk mencari presensi pegawai yang berfungsi untuk menentukan persen pengurangan TPP ASN berdasarkan indikator presensi tepat waktu.

```python
id_pegawai=pegawai['id_pegawai']
id_pegawai
```

Melakukan analisa terhadap data presensi.

```python
presensi
```

Dari output diatas, masih terdapat output NaN. Oleh karena itu, perlu dilakukan cleaning data dengan menghilangkan baris yang mengandung value NaN.

```python
presensi=presensi.dropna()
presensi
```

Mencari informasi terhadap semua atribut/value yang ada di data presensi.

```python
presensi.info()
```

Dari summary diatas, dapat diketahui bahwa terdapat 32 variable/kolom bertipe data float/integer/angka untuk semua informasi terkait pegawai dan presensi nya, dan 1 atribut String untuk atribut syscreate.

Memisahkan data antara waktu presensi dan kehadiran kerja

```python
waktu_presensi=presensi[['id_presensi','id_pegawai','id_unitkerja','bulan','hari kerja', 'hari libur','tepatwaktu','tl1','tl2','tl3','tl4','tl5','psw1','psw2','psw3','psw4','psw5']]
waktu_presensi
```

Menghitung pengurangan persentasi dari setiap id pegawai. Persentasi didapatkan dari acuan Perwal Kota Yogyakarta No.147 Tahun 2020.

```python
percent=[]
for index,row in waktu_presensi.iterrows():
    persen=(1*row['tl1']+2*row['tl2']+3*row['tl3']+4*row['tl4']+8*row['tl5'])+(1*row['psw1']+2*row['psw2']+3*row['psw3']+4*row['psw4']+8*row['psw5'])
    percent.append(persen)
waktu_presensi['persen']=percent
waktu_presensi=waktu_presensi.sort_values('persen',ascending=False)
waktu_presensi
```

Karena terdapat persentase yang melebihi 100%, maka harus diubah menjadi nilai maksimum yaitu 100%.

```python
waktu_presensi['persen']=waktu_presensi['persen'].apply(lambda x: 100 if x>100 else x)
waktu_presensi
```

Melakukan filtering terhadap data waktu presensi, dimana mencari data presensi yang sesuai dengan data pegawai yang ada pada sebelumnya.

```python
waktu_presensi=waktu_presensi[waktu_presensi['id_pegawai'].isin(id_pegawai)]
waktu_presensi
```

Melakukan kalkulasi kumulatif untuk persentase pengurangan TPP ASN dari aspek presensi untuk setiap id pegawai.

```python
new=waktu_presensi.groupby('id_pegawai')['persen'].mean()
new_index=new.index.tolist()
new_value=new.tolist()
temp={new_index[i]:new_value[i]for i in range(len(new_index))}
df=pd.DataFrame(temp.items())
df.columns=['id_pegawai','persen_pengurangan_waktu']
df
```

Menggabung hasil perhitungan kumulatif persentase ke dalam data pegawai dan menjadikanya sebagai kolom baru.

```python
pegawai=pd.merge(pegawai,df,on='id_pegawai')
pegawai
```

```python
pegawai['Pengurangan TPP Presensi']=pegawai['persen_pengurangan_waktu'].apply(lambda x: False if x==0 else True )
pegawai
```

Melakukan analisa terhadap hasil akhir persentase.

```python
result={'Tidak Mendapat Pengurangan':len(pegawai.loc[(pegawai['Pengurangan TPP Presensi'] ==False)]),'Mendapat Pengurangan':len(pegawai.loc[(pegawai['Pengurangan TPP Presensi']==True)])}
print("Terdapat {} atau sekitar {}% pegawai yang taat presensi dan jam pulang kerja dari total keseluruhan  pegawai di Kota Yogyakarta sehingga tidak mendapatkan pengurangan TPP ASN".format(result['Tidak Mendapat Pengurangan'],result['Tidak Mendapat Pengurangan']/len(pegawai)*100),len(pegawai))
```

Menampilkan hasil persentase dalam diagram agar lebih mudah dipahami dan dilihat.

```python
names = list(result.keys())
values = list(result.values())

plt.pie(values,labels=names,startangle=90, shadow=True,explode=(0.1, 0.1), autopct='%1.2f%%')
plt.title('Diagram Pegawai Daerah Istimewa Yogyakarta Berdasarkan Pengurangan TPP ASN indikator presensi')
plt.axis('equal')
plt.show()
```

Dapat disimpulkan bahwa, masih banyak pegawai di Daerah Istimewa Yogyakarta yang mendapat pengurangan TPP ASN karena tidak tepat waktu saat presensi maupun saat pulang kerja

```python
unit_kerja=pegawai.groupby(['nama unit kerja','id_unitkerja'])['persen_pengurangan_waktu'].mean().nlargest(10)
unit_kerja
```

Dari data diatas, dapat dilihat 10 unit kerja dengan pengurangan TPP ASN tertinggi dari aspek presensi dengan indikator presensi tepat waktu dan pulang sebelum waktu kantor

```python
pegawai.nlargest(n=10, columns=['persen_pengurangan_waktu'])
```

Dari data diatas, dapat dilihat 10 peringkat pegawai dengan  pengurangan TPP ASN tertinggi dari aspek presensi dimana nilai terendahnya dimulai dari 27 %. Dan ternyata terdapat fakta unik lain, yaitu pegawai yang mendapatkan pengurangan persen TPP ASN dari aspek presensi berpendidikan S-2 .

```python
kinerja.info()
```

Karena terlalu banyak informasi yang kurang diperlukan dalam melakukan analisis, maka melakukan drop beberapa kolom data.

```python
delete_column_kinerja=['poin diajukan','poin disetujui by verif','disetujui by sistem','ditolak by verif','ditolak by sistem']
for item in delete_column_kinerja:
    del kinerja[item]
kinerja
```

Melakukan filtering terhadap data kinerja yang sesuai dengan data pegawai yang memiliki id kerja yang sama

```python
kinerja=kinerja[kinerja['id_pegawai'].isin(id_pegawai)]
kinerja
```

Melakukan konversi point menjadi nilai berdasarkan acuan target point kinerja menjadi skala 0:100.

```python
target_kinerja={'1.0':7200,'2.0':6800,'3.0':7200,'4.0':6800,'5.0':5400,'6.0':7200}
nilai=[]
for index,row in kinerja.iterrows():
    for key, value in target_kinerja.items():
        if str(row['bulan'])==key:
            nilai.append(round(row['poin aktivitas diakui']/value*100,2)  )
kinerja['nilai kinerja']=nilai
kinerja
```

Membuat fungsi untuk melakukan mapping nilai kinerja agar dapat menentukan persen pengurangan TPP ASN.

```python
def persenKinerja(x):
    if x>=85 and x<=100:
        return 0
    elif x>=75 and x<85:
        return 20
    elif x>=65 and x<75:
        return 30
    elif x>=55 and x<65:
        return 40
    else:
        return 80
```

Konversi nilai kinerja menjadi  pengurangan persen TPP ASN indikator kinerja.

```python
kinerja['persen kinerja']=kinerja['nilai kinerja'].apply(persenKinerja)
kinerja
```

Mencari rata-rata dari persen pengurangan kinerja tiap bulan.

```python
persen_kinerja=kinerja.groupby('id_pegawai')['persen kinerja'].mean()
new_index=persen_kinerja.index.tolist()
new_value=persen_kinerja.tolist()
temp={new_index[i]:new_value[i]for i in range(len(new_index))}
df=pd.DataFrame(temp.items())
df.columns=['id_pegawai','persen_pengurangan_kinerja']
df
```

Menggabung data persen pengurangan TPP ASN indikator kinerja dengan data pegawai.

```python
pegawai=pd.merge(pegawai,df,on='id_pegawai')
pegawai
```

```python
pegawai['Pengurangan TPP Kinerja']=pegawai['persen_pengurangan_kinerja'].apply(lambda x: False if x==0 else True )
pegawai
```

```python
result={'Tidak Mendapat Pengurangan':len(pegawai.loc[(pegawai['Pengurangan TPP Kinerja'] ==False)]),'Mendapat Pengurangan':len(pegawai.loc[(pegawai['Pengurangan TPP Kinerja']==True)])}
print("Terdapat {} atau sekitar {}% pegawai yang kinerja nya sesuai dengan target dari total keseluruhan {} pegawai di Daerah Istimewa Yogyakarta sehingga tidak mendapatkan pengurangan TPP ASN".format(result['Tidak Mendapat Pengurangan'],result['Tidak Mendapat Pengurangan']/len(pegawai)*100,len(pegawai)))
```

Dari data diatas, dapat dikatakan bahwa sebagian besar(3/4) pegawai yang ada di Daerah Istimewa Yogyakarta bekerja sesuai dengan target. Hal ini bisa diketahui dari rata-rata point yang diraih pegawai tiap bulannya. Akan tetapi, masih ada 1/4 pegawai di Kota Yogyakarta yang mendapat pengurangan TPP ASN karena kinerjanya tidak sesuai dengan target yang diindikatorkan melalui point per-bulan.
